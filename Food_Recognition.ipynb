```python
IMGAE_SIZE = [224, 224]
train_path = '/content/drive/MyDrive/Food Classification/train'
valid_path = '/content/drive/MyDrive/Food Classification/val'
test_path  = '/content/drive/MyDrive/Food Classification/test'
```

## **Inceptionv3 model**

```python
inceptionv3 = InceptionV3(input_shape=IMGAE_SIZE + [3], weights='imagenet', include_top=False)
```

```python
for layer in inceptionv3.layers:
  layer.trainble = False
```

```python
num_classes = 20
```

```python
x = Flatten()(inceptionv3.output)
```

```python
prediction = Dense(num_classes, activation='softmax')(x)

# Create a model object 
model_inception = Model(inputs=inceptionv3.input, outputs=prediction)
```

```python
model_inception.summary()
```

```python
model_inception.compile(
    loss='categorical_crossentropy',
    optimizer='adam', 
    metrics=['accuracy']
)
```

## **CNN model**

```python
model_cnn = tensorflow.keras.models.Sequential()
model_cnn.add(layers.Conv2D(32, 3, 1, padding="valid", activation='relu', input_shape=(224, 224, 3)))
model_cnn.add(layers.MaxPool2D(2))
model_cnn.add(layers.Conv2D(64, 3, activation='relu'))
model_cnn.add(layers.MaxPool2D(2))
model_cnn.add(layers.Conv2D(64, 3, activation='relu'))
model_cnn.add(layers.MaxPool2D(2))
model_cnn.add(layers.Conv2D(64, 3, activation='relu'))
model_cnn.add(layers.MaxPool2D(2))
model_cnn.add(layers.Flatten())
model_cnn.add(layers.Dense(128, activation='relu'))
model_cnn.add(layers.Dense(20))
print(model_cnn.summary())
```

```python
model_cnn.compile(
    loss='categorical_crossentropy',
    optimizer='adam', 
    metrics=['accuracy']
)
```

## **Resnet50 model**

```python
resnet = ResNet50(input_shape=IMGAE_SIZE + [3], weights='imagenet', include_top=False)
```

```python
x_resnet = Flatten()(resnet.output)
prediction_resnet = Dense(num_classes, activation='softmax')(x_resnet)

# Create a model object 
model_resnet = Model(inputs=resnet.input, outputs=prediction_resnet)
```

```python
model_resnet.summary()
```

```python
model_resnet.compile(
    loss='categorical_crossentropy',
    optimizer='adam', 
    metrics=['accuracy']
)
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale = 1./255)
val_datagen  = ImageDataGenerator(rescale = 1./255)
```

```python
training_set = train_datagen.flow_from_directory(train_path,
                                                 target_size = (224, 224),
                                                 batch_size = 32,
                                                 class_mode = 'categorical')

test_set = test_datagen.flow_from_directory(test_path,
                                            target_size = (224, 224),
                                            batch_size = 32,
                                            class_mode = 'categorical'
                                            )
val_set = val_datagen.flow_from_directory(valid_path,
                                            target_size = (224, 224),
                                            batch_size = 32,
                                            class_mode = 'categorical'
                                            )
```

```python
r_inception = model_inception.fit_generator(
    training_set,
    validation_data = val_set,
    epochs = 20,
    steps_per_epoch = len(training_set),
    validation_steps = len(val_set)
)
```

```python
r_cnn = model_cnn.fit_generator(
    training_set,
    validation_data = val_set,
    epochs = 20,
    steps_per_epoch = len(training_set),
    validation_steps = len(val_set)
)
```

```python
r_resnet50 = model_inception.fit_generator(
    training_set,
    validation_data = val_set,
    epochs = 20,
    steps_per_epoch = len(training_set),
    validation_steps = len(val_set)
)
```
